# Simulate closed-loop
After implementing your control functionality, you can simulate the closed-loop with code that looks something like this...

```python
# Import relevant modules
from uuv_mission.control import PDController
from uuv_mission.dynamic import ClosedLoop, Mission, Submarine
sub = Submarine()
# Instantiate your controller (depending on your implementation)
controller = PDController(kp=0.15,kd=0.6)
closed_loop = ClosedLoop(sub, controller)
mission = Mission.from_csv("/Users/timotheemahamoodally/b1-coding-practical-mt24/data/mission.csv") # You must implement this method in the Mission clas
trajectory = closed_loop.simulate_with_random_disturbances(mission)
trajectory.plot_completed_mission(mission)
```

```python
import numpy as np

sub = Submarine()

# Initializing gains
best_kp, best_kd = 0, 0
min_error = float('inf')

# Load mission data
mission = Mission.from_csv("/Users/timotheemahamoodally/b1-coding-practical-mt24/data/mission.csv")  # You must implement this method in the Mission class

# Iterate over kp and kd values to find the best combination that gives the lowest mean absolute error between actual and reference trajectory
# Consider values close to kp = 0.15 and kd = 0.6
for kp in np.arange(0.05, 0.3, 0.05):
    for kd in np.arange(0.4, 0.8, 0.05):
        
        sub.reset_state()
        
        # Create a new controller with current kp and kd
        controller = PDController(kp=kp, kd=kd)
        closed_loop = ClosedLoop(sub, controller)

        trajectory = closed_loop.simulate_with_random_disturbances(mission)

        actual_depth = trajectory.position[:, 1]  # Extract the depth from the trajectory
        reference_depth = mission.reference       # The reference trajectory 

        print(f"Kp = {kp}, Kd = {kd}")

        # Find the average error between actual trajectory and reference trajectory
        average_error = np.mean(np.abs(reference_depth - actual_depth))
        if average_error < min_error:
            min_error = average_error
            best_kp, best_kd = kp, kd
        
        print(f"Average error between actual and reference trajectories: {average_error}")
        
        trajectory.plot_completed_mission(mission)

print(f"Optimal combination of Kp and Kd: Kp = {best_kp}, Kd = {best_kd}")
```

```python
sub = Submarine()

controller_optimal = PDController(kp=best_kp, kd=best_kd)  # Optimal gains
controller_initial = PDController(kp=0.15, kd=0.6)         # Initial gains

closed_loop_optimal = ClosedLoop(sub, controller_optimal)
closed_loop_intial = ClosedLoop(sub, controller_initial)

mission = Mission.from_csv("/Users/timotheemahamoodally/b1-coding-practical-mt24/data/mission.csv")

trajectory_optimal = closed_loop_optimal.simulate_with_random_disturbances(mission)
actual_depth_optimal = trajectory_optimal.position[:, 1]  

sub.reset_state()

trajectory_initial = closed_loop_intial.simulate_with_random_disturbances(mission)
actual_depth_initial = trajectory_initial.position[:, 1]  

reference_depth = mission.reference  

print(f"Optimal values of Kp and Kd : Kp = {best_kp}, Kd = {best_kd}, Average error = {np.mean(np.abs(reference_depth - actual_depth_optimal))}")
trajectory_optimal.plot_completed_mission(mission)

print(f"Initial values of Kp and Kd : Kp = 0.15, Kd = 0.6, Average error = {np.mean(np.abs(reference_depth - actual_depth_initial))}")
trajectory_initial.plot_completed_mission(mission)
```
